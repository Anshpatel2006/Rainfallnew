```python
# Import libraries
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv("austin_weather.csv")
data
```

```python

# Drop unnecessary columns
data = data.drop(["Events", "Date", "SeaLevelPressureLowInches"], axis=1)
```

```python

# Replace 'T' and '-' with 0.0
data = data.replace(['T', '-'], 0.0)
```

```python
data
```

```python
# Save cleaned data
data.to_csv("austin_weather_final.csv", index=False)
```

```python

# Reload cleaned data
data = pd.read_csv("austin_weather_final.csv")
data
```

```python

X = data.drop(['PrecipitationSumInches'], axis=1)
```

```python
Y = data["PrecipitationSumInches"]
Y = Y.values.reshape(-1, 1)
Y
```

```python
day_index = 798
days = [ i for i in range(Y.size) ]
```

```python
# Make predictions
clf = LinearRegression()
clf.fit(X, Y)
```

```python
# plot a graph
print("The precipitation trend graph:")

plt.scatter(days, Y, color='g')
plt.scatter(days[day_index], Y[day_index], color='r')
plt.title("Precipitation level")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")
plt.show()

x_vis = X.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent',
                  'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
print("The precipitation vs attributes trend graph:")

for i in range(x_vis.columns.size):
    plt.subplot(3, 2, i+1)
    plt.scatter(days[:800], x_vis[x_vis.columns.values[i]][:800], color='g')
    plt.scatter(days[day_index], x_vis[x_vis.columns.values[i]][day_index], color='r')
    plt.title(x_vis.columns.values[i])

plt.show()
```
